```python
def isValid(s:str) -> bool:
    hash_val = {
        ')': '(',
        ']': '[', 
        '}': '{'
        }
    stack = []
    for i in range(len(s)):
        if s[i] in hash_val:
            if stack and stack[-1] == hash_val[s[i]]:
                stack.pop()
            else:
                return False
        else:
            stack.append(s[i])
    return not stack

isValid('()[]{}')                   
```

```python
def romantoint(s:str) -> int:
    hash_val = {
        'I' : 1,
        'V' : 5,
        'X' : 10,
        'L' : 50,
        'C' : 100,
        'D' : 500,
        'M' : 1000
    }

    n = len(s)
    res = 0

    for i in range(n):
        if  i + 1 < n and hash_val[s[i]] < hash_val[s[i+1]]:
            res-=hash_val[s[i]]
        else:
            res+=hash_val[s[i]]
    return res
```

```python
def intToRoman(num:int)->str:
    
    symbols = ['M','CM','D','CD','C','XC','L','XL','X','IX','V','IV','I']
    values =  [1000,900,500,400,100,90,50,40,10,9,5,4,1]
    res = ""
    for i in range(len(values)):
        count = num // values[i]
        res += symbols[i] * count
        num %= values[i]
    return res
intToRoman(3749)
```

## Sorting algorithms

```python
# 1)selection sort


def selection_sort(nums:list[int])-> list[int]:
    """ Here in this algo first select the min value and 
       swap it with the min value variable.

    Args:
        nums (list[int]): here it take a list of integers

    Returns:
        list[int]: sorted inplace list
    """
    n = len(nums)
    for i in range(n):
        min_indx = i
        for j in range(i+1,n):
            if nums[min_indx] > nums[j]:
                min_indx = j
        nums[min_indx],nums[i] = nums[i],nums[min_indx]
    return nums
```

```python
selection_sort([3,4,2,1,5,7,6,8])
```

## Bubble sort

```python
def bubble_sort(nums:list[int]) -> list[int]:
    """It is also calledd as adjacent swapping.

    Args:
        nums (list[int]): here it take a list of integers

    Returns:
        list[int]: sorted inplace list
    """
    n = len(nums)
    is_swapped = False
    for i in range(n-2,-1,-1):
        for j in range(0,i+1):
            if nums[j] > nums[j + 1]:
                nums[j],nums[j+1] = nums[j+1],nums[j]
                is_swapped = True
        if not is_swapped:
            break
            
    return nums
```

```python
bubble_sort([3,4,2,1,5,7,6,8])
```

## Insertion sort

```python
def insertion_sort(nums:list[int]) -> list[int]:
    n = len(nums)
    for i in range(1,n):
        key = nums[i]
        j = i-1
        while j >= 0 and nums[j] > key:
            nums[j+1] = nums[j]
            j-=1    
        nums[j+1] = key
        
    return nums
```

```python
insertion_sort([3,4,2,1,5,7,6,8])
```

## Merge sort

```python
def merge_array(left:list[int],right:list[int]) -> list[int]:
    i,j = 0,0
    n,m = len(left),len(right)
    result = []
    
    while i<n and j<m:
        if left[i] <= right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
    while i<n:
        result.append(left[i])
        i+=1
    while j<m:
        result.append(right[j])
        j+=1
    return result
```

```python
def merge_sort(nums:list[int]) -> list[int]:
    if len(nums)<=1:
        return nums
    mid = len(nums)//2
    left_arr = nums[:mid]
    right_arr = nums[mid:]
    
    left = merge_sort(left_arr)
    right = merge_sort(right_arr)
    
    return merge_array(left,right)
```

```python
merge_sort([38, 27, 43, 3, 9, 82, 10])
```

## Quick sort

```python
#step1 partition

def partition(nums:list[int],low:int,high:int) -> int:
    pivot = nums[0]
    i = low
    j = high
    while i<j:
        while nums[i] <= pivot and i<= high-1:
            i+=1
        while nums[j] >= pivot and j >= low + 1:
            j-=1
        if i<j:
            nums[i],nums[j] = nums[j],nums[i]
    nums[low],nums[j] = nums[j],nums[low]
    return j
```

```python
def quick_sort(nums:list[int],low:int,high:int) -> int:
    if low<high:
        p_idx = partition(nums,low,high)
        quick_sort(nums,low,p_idx-1)
        quick_sort(nums,p_idx+1,high)
    return nums
```

```python
nums = [4,1,7,6,3,2,8]
low = 0
high = len(nums) - 1
quick_sort(nums,low,high)
```

```python
def largest_num(nums:list[int]) -> int:
    if not nums:
        raise ValueError('input list cannot be empty')
    largest = nums[0]
    n = len(nums)
    for i in range(1,n):
        largest = max(nums[i],largest)
    return largest
```

```python
largest_num([55,32,-97,3,67])
```

```python
def second_largest_num(nums:list[int]) -> int:
    largest = float('-inf')
    s_largest = float('-inf')
    n = len(nums)
    for i in range(n):
        if nums[i] > largest:
            s_largest = largest
            largest = nums[i]
        elif s_largest < nums[i] and s_largest != largest:
            s_largest = nums[i]
    return s_largest
            
```

```python
second_largest_num([55,32,97,-55,45,32,88,21])
```

```python
def check_sort(nums:list[int]) -> bool:
    n = len(nums)
    
    for i in range(n-1):
        if nums[i] >= nums[i+1]:
            return False
    return True
```

```python
check_sort([1,2,4,3,5,6])
```

```python
def unique_elements(nums:list[int]) -> int:
    n = len(nums)
    i = 0
    for j in range(1,n):
        if nums[i] != nums[j]:
            i+=1
            nums[i] = nums[j]
    return i + 1
```

```python
unique_elements([1,1,1,2,3,4,4,7,9,9,9,10])
```

```python
def right_rotate(nums:list[int])->list[int]:
    temp = nums[-1]
    n = len(nums)
    for i in range(n-2,-1,-1):
        nums[i+1] = nums[i]
    nums[0] = temp
    return nums
```

```python
right_rotate([1,2,3,4,5,6,7])
```

```python
def reverse(nums:list[int],left:int,right:int) -> list[int]:
    while left < right:
        nums[left],nums[right] = nums[right],nums[left]
        left+=1
        right-=1
```

```python
def k_rotate(nums:list[int],k:int)->list[int]:
    n=len(nums)
    k %= n
    reverse(nums,n-k,n-1)
    reverse(nums,0,n-k-1)
    reverse(nums,0,n-1)
    return nums
    
```

```python
k_rotate([1,2,3,4,5,6,7],5)
```
